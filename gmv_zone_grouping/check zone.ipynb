```python
# # swap x and y
for n in range(len(df3.area)):
    a = json.loads(df3.area[n])
    
    for i in range(len(a['coordinates'][0])):
        left = a['coordinates'][0][i][0]
        right = a['coordinates'][0][i][1]
        a['coordinates'][0][i][0] = right
        a['coordinates'][0][i][1] = left
    df3.area[n] = shape(a)
```
